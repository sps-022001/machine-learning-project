**MACHINE LEARNING PROJECT TO PREDICT THE TYPE OF WINE**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mainfile=pd.read_csv('/content/WineQuality.csv')
mainfile
```

```python
empty_cells=mainfile.isnull().sum()
empty_cells
```

```python
mainfile=mainfile.dropna()
```

```python
data_type=mainfile.dtypes
print(data_type)
```

```python

statical=mainfile.describe()
statical
```

```python
x=mainfile.drop('Type',axis=1)
```

```python
y=mainfile['Type']
```

```python
print(x)
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x.shape,xtrain.shape,xtest.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
xtrain_prediction=model.predict(xtrain)
trainingacc=accuracy_score(xtrain_prediction,ytrain)
```

```python
print(trainingacc)
```

```python
xtest_prediction=model.predict(xtest)
tacc=accuracy_score(xtest_prediction,ytest)
print(tacc)
```
